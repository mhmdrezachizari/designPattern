```python
from abc import ABC, abstractmethod


class Observer(ABC):

    @abstractmethod
    def update(self):
        pass


class EmailObserver(Observer):

    def update(self):
        print("Email Sent")


class SmsObserver(Observer):

    def update(self):
        print("SMS Sent")


class LogObserver(Observer):

    def update(self):
        print("Log Created")


class Order:

    def __init__(self):
        self.observers = []

    def attach(self, observer):
        self.observers.append(observer)

    def notify(self):
        for observer in self.observers:
            observer.update()

    def create(self):
        print("Order Created")
        self.notify()


order = Order()

order.attach(EmailObserver())
order.attach(SmsObserver())
order.attach(LogObserver())

order.create()
```
