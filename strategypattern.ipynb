```python
from abc import ABC, abstractmethod



class DiscountStrategy(ABC):

    @abstractmethod
    def calculate(self, price: int) -> int:
        pass


class VipDiscount(DiscountStrategy):

    def calculate(self, price: int) -> int:
        return int(price * 0.9)


class StudentDiscount(DiscountStrategy):

    def calculate(self, price: int) -> int:
        return int(price * 0.85)


class FixedDiscount(DiscountStrategy):

    def calculate(self, price: int) -> int:
        return max(price - 100_000, 0)


class NoDiscount(DiscountStrategy):

    def calculate(self, price: int) -> int:
        return price


class DiscountFactory:

    _strategies = {
        "vip": VipDiscount,
        "student": StudentDiscount,
        "fixed": FixedDiscount,
        "none": NoDiscount,
    }

    @staticmethod
    def create(name: str) -> DiscountStrategy:
        strategy_class = DiscountFactory._strategies.get(name.lower())

        if not strategy_class:
            raise ValueError(f"Unknown discount strategy: {name}")

        return strategy_class()


class Order:

    def __init__(self, price: int, strategy: DiscountStrategy):
        self.price = price
        self.strategy = strategy

    def final_price(self) -> int:
        return self.strategy.calculate(self.price)




```

```python
price = 1_000_000

order1 = Order(price, VipDiscount())
print("VIP:", order1.final_price())

strategy = DiscountFactory.create("student")
order2 = Order(price, strategy)
print("Student:", order2.final_price())

order3 = Order(price, DiscountFactory.create("fixed"))
print("Fixed:", order3.final_price())

order4 = Order(price, DiscountFactory.create("none"))
print("None:", order4.final_price())
```
