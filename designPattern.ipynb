Singleton


```python
class Singleton:

    _instance = None

    def __new__(cls):

        if cls._instance is None:
            print("Creating instance...")

            cls._instance = super().__new__(cls)

        return cls._instance
```

```python
db1 = Singleton()
db2 = Singleton()
print(db1 is db2)
```

```python
db3 = Singleton()
print(db1 is db3)
```

Factory


```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


class ZarinPal(PaymentGateway):

    def pay(self, amount):
        print(f"✅ Paying {amount:,} Toman via ZarinPal")


class IDPay(PaymentGateway):

    def pay(self, amount):
        print(f"✅ Paying {amount:,} Toman via IDPay")


class NextPay(PaymentGateway):

    def pay(self, amount):
        print(f"✅ Paying {amount:,} Toman via NextPay")


# Factory
class PaymentFactory:

    @staticmethod
    def create(gateway_name):

        gateways = {
            "zarinpal": ZarinPal,
            "idpay": IDPay,
            "nextpay": NextPay,
        }

        gateway = gateways.get(gateway_name.lower())

        if gateway is None:
            raise ValueError(
                f"Gateway '{gateway_name}' not found"
            )

        return gateway()



gateway_name = "zarinpal"

payment_gateway = PaymentFactory.create(gateway_name)

print(type(payment_gateway))

payment_gateway.pay(500000)
```
