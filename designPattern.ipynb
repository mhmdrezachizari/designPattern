```python
class Singleton:

    _instance = None

    def __new__(cls):

        if cls._instance is None:
            print("Creating instance...")

            cls._instance = super().__new__(cls)

        return cls._instance
```

```python
db1 = Singleton()
db2 = Singleton()
print(db1 is db2)
```

```python
db3 = Singleton()
print(db1 is db3)
```
